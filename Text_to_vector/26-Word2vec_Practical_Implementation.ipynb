# Word2vec

Word2vec is a technique in natural language processing (NLP) for obtaining vector representations of words. These vectors capture information about the meaning of the word based on the surrounding words. The word2vec algorithm estimates these representations by modeling text in a large corpus. Once trained, such a model can detect synonymous words or suggest additional words for a partial sentence.

```python
# using a pretrained word2vec model from google
import gensim.downloader as api
# loading the model from google
wv = api.load('word2vec-google-news-300')
# checking the vector value of word 'King'
vec_king = wv['king']
```

```python
wv['cricket']
```

```python
# similar words to the word 'cricket'
# the model convert 'cricket' to vector and  chek the nearby vectors and convert them to words
wv.most_similar('cricket')
```

```python
wv.most_similar('happy')
```

```python
# similarity of two words
# cosine similarity is used to calculate the similarity between two words
wv.similarity("hockey","sports")
```

```python
# king - man + women = Queen
vec=wv['king']-wv['man']+wv['woman']
```

```python
vec
```

```python
# here we check the simar words, we can see queen.
wv.most_similar([vec])
```
